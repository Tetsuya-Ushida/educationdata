```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = np.loadtxt("section5-2.csv",delimiter=",",skiprows=1,usecols=(1,2),unpack=True)
```

```python
np.corrcoef(data[0],data[1])
```

```python
a = np.cov(data[0],data[1])[0][1]/np.var(data[0])
b = np.average(data[1])-a*np.average(data[0])
```

```python
y_h = a*data[0]+b
```

```python
y_h
```

```python
plt.figure()
plt.scatter(data[0],data[1])
plt.plot(data[0],y_h,color="orange")
```
